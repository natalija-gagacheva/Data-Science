# 1. Handling missing values

1. **Pairwise deletions** - when values are missing completely at random (MCAR), they are deleted.
2. **Listwise/Dropping entire rows** - deleting the row containing the missing value. Problem: a major chunk of data is lost, use only for a small amount of lost data
4. **Dropping entire columns** - deleting the column containing the missing value if said column contains a lot of missing values and the feature isn't necessarily significant

```python
podatoci.head(10)
```

**Listwise/Dropping entire rows** - .dropna() method removes rows where there are missing (NaN) values. Axis=0 specifies that rows should be dropped.

```python
podatoci = podatoci.dropna(subset = ['FoodCourt'], axis = 0)
```

```python
podatoci.head(10)
```

**Notice that in both RoomService and FoodCourt, the sum is 0 since we deleted the values**

# 2. Data Imputation

1. **General** - Imputing with a constant, mean, meadian

```python
podatoci.head(10)
```

**Constant Imputer**

In this case, SimpleImputer allows you to replace missing values with a specified constant value = 0. This is a common imputation approach when missing values are assumed to mean "no data" or "no usage"

```python
constantImputer = SimpleImputer (strategy = 'constant', fill_value=0)
```

```python
# .fit_transform() replaces all NaN values with the constant 0
```

```python
podatoci['FoodCourt'] = constantImputer.fit_transform(podatoci[['FoodCourt']])
```

```python
podatoci.isnull().sum().sort_values(ascending=False)
```

**Frequent Imputer**

```python
frequentImputer = SimpleImputer(strategy = 'most_frequent')
```

```python
podatoci['HomePlanet'] = podatoci['HomePlanet'].astype(str)                                                
```

# IMPORTANT: .fit_transform() expects a 2D array as input argument

```python
podatoci['HomePlanet'] = frequentImputer.fit_transform(podatoci['HomePlanet'].values.reshape(1,-1))[0]
```

What each part of this code does:
1. **podatoci['HomePlanet']** --- selects that column from the DataFrame
2. **podatoci['HomePlanet'].values** --- extracts and converts that column to a 1D array
3. **.reshape (1, -1)** --- reshapes the 1D array into a 2D array with one row and as many columns as there are elements in HomePlanet. Example:
    > 1. 1D array: ages = [23, 34, 45, 56 ]
    > 2. 2D array: info = [['Alice', 23], ['Emma', 45], ['Nick', 19]]
4. **frequentImputer.fit_transform()** --- replaces missing values with the most frequent value
5. **[0]** --- extracts the 0th row from the 2D array, converting it back to a 1D array

```python
podatoci.isnull().sum().sort_values()
```

```python
podatoci.Age = podatoci.Age.fillna(podatoci.Age.mean())
```

```python
podatoci.isnull().sum().sort_values()
```

```python
podatoci.ShoppingMall.mean()
```

```python
podatoci.ShoppingMall.median()
```

```python
podatoci.ShoppingMall = podatoci.ShoppingMall.fillna(podatoci.ShoppingMall.median())
```

What each part of this code does:
1. podatoci.ShoppingMall --- access that column
2. **.fillna()** --- fills the missing values with the column median (in this case, 0.0)

1. .bfill() fills the missing values with the first valid value that comes **after** each missing value
   > aq_data['NO2'] = aq_data['NO2'].bfill()
3. .ffill() fills the missing values with the most recent non-missing value that comes **before** in the column
    > aq_data['CO'] = aq_data['CO'].ffill()


3. interpolation method estimates missing values based on the surrounding non-missing values. It creates a straight line between the last valid observation before the missing value and the next valid observation after the missing value. the limit_direction='both' ensures that missing values at the beginning and end of the series can also be filled
   >aq_data['PM10'] = aq_data['PM10'].interpolate(limit_direction='both')

# Custom Encoding

```python
def enkodirajBoolVariable(x):
    if type(x) == float:
        return x
    return 1 if x else 0
```
